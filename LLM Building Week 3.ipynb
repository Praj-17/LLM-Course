## Tokenization
Tokenization is the process of breaking down text into smaller units, known as tokens. Tokens can be words, characters, or subwords. These tokens are the building blocks for methods like Bag of Words, TF-IDF, and word embeddings. Let's explore common tokenization techniques.

### Basic Word Tokenization
We'll begin with a simple text example and use NLTK, a popular Natural Language Toolkit, to tokenize the text into words.

```python
from nltk.tokenize import word_tokenize

text = "Tokenization is the process of converting text into tokens."
word_tokens = word_tokenize(text)
print("Example 1:", word_tokens)

# Second example: punctuation and contractions
text2 = "I'm learning NLP—it's fascinating!"
word_tokens2 = word_tokenize(text2)
print("Example 2:", word_tokens2)
```

### Bag of Words (BOW)
A simple representation that counts words but ignores order. Each document is a vector of word counts over a fixed vocabulary.

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

documents = [
    "The cat sat on the mat.",
    "The dog sat on the log.",
    "The cat and the dog are friends.",
]
cv = CountVectorizer()
count_matrix = cv.fit_transform(documents)
print("Vocabulary:", cv.get_feature_names_out())
print("Shape:", count_matrix.toarray().shape)
df_bow = pd.DataFrame(count_matrix.toarray(), columns=cv.get_feature_names_out(), index=["Doc1", "Doc2", "Doc3"])
df_bow
```

### TF-IDF (Term Frequency-Inverse Document Frequency)
Weights word frequency by rarity across documents: terms that appear in many documents get lower weight; terms specific to a document get higher weight.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)
df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray().round(3),
    columns=tfidf.get_feature_names_out(),
    index=["Doc1", "Doc2", "Doc3"]
)
df_tfidf
# Compare: "the" appears in all docs (low IDF) vs "mat" only in Doc1 (high IDF)
print("\nTF-IDF for 'the' in Doc1:", tfidf_matrix[0, tfidf.vocabulary_['the']])
print("TF-IDF for 'mat' in Doc1:", tfidf_matrix[0, tfidf.vocabulary_['mat']])
```

### Word2Vec
Transforms words into dense high-dimensional vectors that capture context; similar words have similar vectors. We train a small model on a few sentences for illustration.

```python
from gensim.models import Word2Vec
sentences = [
    ["nlp", "is", "fun"],
    ["machine", "learning", "is", "great"],
    ["nlp", "uses", "machine", "learning"],
]
model_w2v = Word2Vec(sentences, vector_size=10, window=2, min_count=1, seed=42)
print("Vector for 'nlp' shape:", model_w2v.wv["nlp"].shape)
print("Words most similar to 'nlp':", model_w2v.wv.most_similar("nlp", topn=3))
```

### Advanced Tokenization with Transformers
For complex models, tokenization needs to preserve contextual information. The Transformers library provides advanced tokenization tools.

```python
from transformers import BertTokenizer

# Initializing the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenizing a sample text
sample_text = "Transformers have revolutionized NLP."
input_ids = tokenizer.encode(sample_text, add_special_tokens=True)
tokens = tokenizer.convert_ids_to_tokens(input_ids)
tokens
```

### Subword Tokenization
BERT uses subword tokenization: rare or long words are split into subword units (e.g. "unhappiness" -> "un", "##happiness"), so the model can handle unseen words and keep vocabulary size manageable.

```python
# Subword breakdown for long/rare words
for word in ["unhappiness", "tokenization", "Transformers"]:
    ids = tokenizer.encode(word, add_special_tokens=False)
    subwords = tokenizer.convert_ids_to_tokens(ids)
    print(f"{word!r} -> {subwords}")
```

## Stemming and Lemmatization
Stemming and Lemmatization are text normalization techniques that reduce words to their base form.
- **Stemming** removes suffixes to bring words to a base form.
- **Lemmatization** considers the context and reduces words to their meaningful base form.

### Stemming Example
We'll use NLTK's Porter Stemmer to demonstrate stemming.

```python
from nltk.stem import PorterStemmer

# Initializing the stemmer
stemmer = PorterStemmer()

# Words to stem
words = ["running", "runs", "runner"]
# Applying stemming
stemmed_words = [stemmer.stem(word) for word in words]
stemmed_words
```

### Lemmatization Example
NLTK's WordNet Lemmatizer provides lemmatization with more linguistic context.

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

# Initializing the lemmatizer
lemmatizer = WordNetLemmatizer()

# Words to lemmatize (with correct POS for better results)
words = ["running", "runs", "runner"]
lemmas_v = [lemmatizer.lemmatize(word, pos='v') for word in words]  # verb
lemmas_n = [lemmatizer.lemmatize(word, pos='n') for word in words]  # noun
print("As verb:", lemmas_v)
print("As noun:", lemmas_n)
```

### Stemming vs Lemmatization: Comparison
Same sentence processed by both techniques. Stemming is faster and order-independent; lemmatization uses context (e.g. POS) and yields valid words.

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

sentence = "The runners are running quickly and the runs were successful."
tokens = word_tokenize(sentence)
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed = [stemmer.stem(w) for w in tokens]
# Lemmatize with default (noun); for "running" we need verb
lemmatized = [lemmatizer.lemmatize(w, pos='v') if w in ('running', 'runs', 'were') else lemmatizer.lemmatize(w) for w in tokens]

print("Original: ", tokens)
print("Stemmed:  ", stemmed)
print("Lemmatized:", lemmatized)
```

### Snowball Stemmer (optional variety)
Snowball supports multiple languages; here we use the English stemmer for comparison with Porter.

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer('english')
words_sample = ["running", "beautiful", "happiness"]
print("Porter:  ", [stemmer.stem(w) for w in words_sample])
print("Snowball:", [snowball.stem(w) for w in words_sample])
```

## Text-Level Applications
Techniques such as stop-word removal, part-of-speech tagging, and normalization are critical in preprocessing text for analysis and machine learning.

### Stop-Word Removal
Removing common words (e.g. "the", "is") that often add little signal for many NLP tasks.

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
paragraph = "The quick brown fox jumps over the lazy dog. It is a sunny day and the birds are singing."
tokens_para = word_tokenize(paragraph.lower())
filtered = [w for w in tokens_para if w.isalnum() and w not in stop_words]
print("Before:", tokens_para)
print("After: ", filtered)
print("Sample stop words:", list(stop_words)[:15])
```

### Part-of-Speech Tagging
POS tags help with lemmatization, feature engineering, and understanding sentence structure.

```python
from nltk import pos_tag
sentence_pos = "The runners are running quickly."
tokens_pos = word_tokenize(sentence_pos)
tagged = pos_tag(tokens_pos)
print("Word -> POS tag")
for word, tag in tagged:
    print(f"  {word:12} -> {tag}")
```

### Normalization
Lowercasing and removing punctuation (and optionally Unicode normalization) make text consistent for downstream models.

```python
import re
import unicodedata
raw = "Hello! This is an Example—with mixed CASE and punctuation."
normalized = raw.lower()
normalized = re.sub(r'[^\w\s]', '', normalized)  # remove punctuation
normalized = re.sub(r'\s+', ' ', normalized).strip()
print("Before:", raw)
print("After: ", normalized)
# Optional: Unicode normalization (e.g. normalize accented characters)
sample_unicode = "café"
print("NFC:", unicodedata.normalize('NFC', sample_unicode), "NFD:", unicodedata.normalize('NFD', sample_unicode))
```

## Introduction to Transformers
Transformers are advanced neural network architectures designed for processing sequential data efficiently. They use **self-attention** to let each token attend to all other tokens, capturing long-range context and improving NLP tasks. Unlike older recurrent models, they allow parallel computation and have become the backbone of models like BERT, GPT, and T5.

### Self-Attention Intuition
In self-attention, each token computes a weighted sum over all tokens (including itself). The weights indicate "how much to attend" to each position. Below is a minimal hand-crafted example: 3 tokens with a 3x3 attention weight matrix.

```python
import numpy as np
# Example: 3 tokens, attention weights (each row sums to 1)
# Token 0 attends mostly to itself and token 1; token 2 attends to all
attention_weights = np.array([
    [0.6, 0.3, 0.1],  # token 0
    [0.2, 0.5, 0.3],  # token 1
    [0.33, 0.33, 0.34],  # token 2
])
print("Attention weight matrix (rows = from, cols = to):")
print(attention_weights)
print("Each row sums to 1:", attention_weights.sum(axis=1))
```

### BERT in Action: Masked Language Model
BERT uses bidirectional context; it can fill in a masked token using both left and right context. Below we use the fill-mask pipeline to predict the masked word.

```python
fill_mask = pipeline("fill-mask", model="bert-base-uncased")
result = fill_mask("The capital of France is [MASK].")
for r in result:
    print(f"  {r['token_str']!r}: score = {r['score']:.4f}")
```

BERT uses bidirectional context and is well-suited for understanding nuances (e.g. sentiment, NLI). For classification tasks, pre-trained BERT is often fine-tuned on labeled data.

### Using Models from Hugging Face
Hugging Face provides interfaces to easily integrate complex models into applications. Let's implement a basic usage of one of their models.

```python
from transformers import BertTokenizer, GPT2Tokenizer, RobertaTokenizer
# Tokenizers only (no full model) to show model availability
models = [
    ("bert-base-uncased", BertTokenizer),
    ("gpt2", GPT2Tokenizer),
    ("roberta-base", RobertaTokenizer),
]
for name, tokenizer_class in models:
    tok = tokenizer_class.from_pretrained(name)
    print(f"Loaded: {name}")
```

```python
from transformers import pipeline

# Using a pipeline for sentiment analysis
nlp_pipeline = pipeline("sentiment-analysis")
result = nlp_pipeline("Hugging Face is transforming NLP!")
print("Example 1:", result)
# Multiple sentences
sentences = ["I love this course!", "This is terrible.", "It's okay, nothing special."]
for sent, res in zip(sentences, nlp_pipeline(sentences)):
    print(f"  {sent!r} -> {res['label']} ({res['score']:.3f})")
```

### More Usage Examples
Below: Named-entity recognition and question answering using Hugging Face pipelines.

### Named-Entity Recognition (NER)
Extract named entities (persons, organizations, locations, etc.) from text.

```python
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")
text_ner = "Hugging Face is based in New York and was founded by Clement Delangue."
entities = ner_pipeline(text_ner)
for e in entities:
    print(f"  {e['word']!r}: {e['entity_group']} (score={e['score']:.3f})")
```

### Question Answering
Answer a question given a context passage.

```python
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilledsquad")
context = "Hugging Face is a company that provides NLP tools and models. It is headquartered in New York."
answer = qa_pipeline(question="Where is Hugging Face headquartered?", context=context)
print(f"Answer: {answer['answer']!r} (score={answer['score']:.3f})")
```
